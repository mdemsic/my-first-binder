```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, FloatSlider

def displacement(t, A, B, omega):
    return A * np.cos(omega * t) + B * np.sin(omega * t)

def plot_vibration(A, B, omega):
    t = np.linspace(0, 10, 400)  # time from 0 to 10 seconds, 400 points
    x = displacement(t, A, B, omega)

    plt.figure(figsize=(10, 6))
    plt.plot(t, x)
    plt.title('Free Undamped Vibrations')
    plt.xlabel('Time (s)')
    plt.ylabel('Displacement (m)')
    plt.grid(True)
    
    # Set fixed y-axis limits
    plt.ylim(-3, 3)  # Adjust these values as needed for your specific visualization

    plt.show()

interactive_plot = interactive(plot_vibration,
                               A=FloatSlider(value=1.0, min=-2.0, max=2.0, step=0.1, description='Initial Displacement (A)'),
                               B=FloatSlider(value=0.0, min=-2.0, max=2.0, step=0.1, description='Initial Velocity (B)'),
                               omega=FloatSlider(value=2*np.pi, min=0, max=10, step=0.1, description='Natural Frequency (ω)'))
interactive_plot
```
